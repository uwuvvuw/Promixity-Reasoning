# Dataset Preparation

```python
import pandas as pd
import numpy as np
```

```python
# constructing the dataframe
# dataset is about whether a person is actually under the category 'sus'

df = pd.DataFrame({'GENDER': {0: 'M',
  1: 'F',
  2: 'M',
  3: 'M',
  4: 'F',
  5: 'F',
  6: 'M',
  7: 'F',
  8: 'M',
  9: 'F'},
 'MONTH': {0: 'JAN',
  1: 'FEB',
  2: 'JAN',
  3: 'MARCH',
  4: 'FEB',
  5: 'JUNE',
  6: 'AUGUST',
  7: 'FEB',
  8: 'JAN',
  9: 'FEB'},
 'CITY': {0: 'KL',
  1: 'KEDAH',
  2: 'SELANGOR',
  3: 'PINENG',
  4: 'PERLIS',
  5: 'JOHOR',
  6: 'KEDAH',
  7: 'SELANGOR',
  8: 'KL',
  9: 'PERAK'},
 'OCCUPATION': {0: 'MANAGER',
  1: 'DOCTOR',
  2: 'EXECUTIVE',
  3: 'CLERK',
  4: 'LECTURER',
  5: 'SUPERVISOR',
  6: 'CLEANER',
  7: 'TAILOR',
  8: 'TEACHER',
  9: 'PILOT'},
 'SUSPICIOUS': {0: 'Y',
  1: 'Y',
  2: 'N',
  3: 'Y',
  4: 'N',
  5: 'N',
  6: 'N',
  7: 'Y',
  8: 'N',
  9: 'Y'}})

df
```

# Dummification

```python
# dummifying the variables
df_all = df.drop('SUSPICIOUS',axis=1)
df_all = pd.get_dummies(df_all)
df_all
```

# New Customer

```python
new_customer = pd.DataFrame({
    'GENDER':['F'],
    'MONTH':['FEB'],
    'CITY':['KL']
})
new_customer
```

```python
# why dummification? bec we need to prepare a dataset
pd.get_dummies(new_customer)
```

# Data Transformation (New Customer)

```python
value = list()
for i in range(0,len(new_customer.columns)):    
    str = new_customer.columns[i] + '_' + new_customer.iloc[0,i]
    value.append(str)
```

```python
# create a copy of orginal dataframe but with just one record

df_new_customer = df_all.head(1).copy()
df_new_customer
```

```python
# set values of all columns to 0
# intialize everything into 0
df_new_customer.values[:] =0
df_new_customer
```

# Set value 1 to respective columns

```python
# set all the related data to 1
# respective columns will be set to 1
for i in range(0,len(value)):
    df_new_customer.loc[:,value[i]]=1
```

```python
df_new_customer
```

# Perform Similarity Test using cdist()

```python
from scipy.spatial.distance import cdist
# cdist is the best, in terms of large dataset (very fast)
```

```python
# create 3 variables with 3 different metrics - euclidean, hamming, cosine
# cdist against 3 diff metrics

score_euclidean = cdist(df_new_customer, df_all, metric='euclidean')
score_hamming = cdist(df_new_customer, df_all, metric='hamming')
score_cosine = cdist(df_new_customer, df_all, metric='cosine')

# concate back to data frame
df_all['Score_euclidean'] = score_euclidean[0]
df_all['Score_hamming'] = score_hamming[0]
df_all['Score_cosine'] = score_cosine[0]

# euclidean score can go very large number
# the larger the score of euclidean, the larger the distance
# cosine decides the direction (0-1)
```

```python
df_all
```

# Merge back the Suspicious column

```python
df_complete = pd.concat([df_all, df[['SUSPICIOUS']]], axis=1)
df_complete

# do not sort the dataset/dataframe [so we can get back the sus]
```

# Decision Making

```python
# Sort the dataframe by euclidean socre

df_complete = df_complete.sort_values(by=['Score_euclidean'])
df_complete

# the first three rows show the top 3 nearest customers to the new customer
# we can conclude that the new customer has a high probability of being sus
# because there are 2Y and 1N in the top three nearest customers
```

```python
# using the list to append all the decision from each metric.

final_decision_list = []

# euclidean
df_complete = df_complete.sort_values(by=['Score_euclidean'])
df_complete = df_complete.reset_index(drop=True)
decision_list = df_complete['SUSPICIOUS'][0:5].tolist()
if decision_list.count('Y')>decision_list.count('N'):
    final_decision_list.append('Y')
else:
    final_decision_list.append('N')

# hamming
df_complete = df_complete.sort_values(by=['Score_hamming'])
df_complete = df_complete.reset_index(drop=True)
decision_list = df_complete['SUSPICIOUS'][0:5].tolist()
if decision_list.count('Y')>decision_list.count('N'):
    final_decision_list.append('Y')
else:
    final_decision_list.append('N')

# cosine
df_complete = df_complete.sort_values(by=['Score_cosine'])
df_complete = df_complete.reset_index(drop=True)
decision_list = df_complete['SUSPICIOUS'][0:5].tolist()
if decision_list.count('Y')>decision_list.count('N'):
    final_decision_list.append('Y')
else:
    final_decision_list.append('N')

final_decision_list
```

```python
# Build the logic to spell out the final decision.
# That is if hamming->Y, cosine->Y, euclidean->N, the final decision is Y

if final_decision_list.count('Y')>final_decision_list.count('N'):
    print('Final Decision = ' + 'SUSPICIOUS')
else:
    print('Final Decision = ' + 'NOT SUSPICIOUS')
```
